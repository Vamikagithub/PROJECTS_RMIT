```python
#DATA PREPARATION
#importing files using necessary library
import pandas as aut
aut.set_option('display.max_rows', 275)
aut.set_option('display.max_columns', 30)
```

```python
#reading the csv file into Python environment 
auto_filename = 'Automobile.csv'
auto = aut.read_csv(auto_filename,sep='#',decimal='.',header = None,
                   names = ['Symboling','Normalized Losses','Make','Fuel Type','Aspiration','No.of Doors',
                           'Body Style','Drive Wheels','Engine Location','Wheel Base','Length','Width','Height',
                            'Curb Weight','Engine Type','No. of Cylinders','Engine Size','Fuel System','Bore','Stroke',
                           'Compression Ratio','Horsepower','Peak RPM','City MPG','Highway MPG','Price'])
```

```python
#Replacing Model make,Aspiration,No.of Doors with the correct name
auto['Make'].replace('alfa-romero','alfa-romeo',inplace=True)
auto['Make'].replace('Nissan','nissan',inplace=True)
auto['Make'].replace('volvo ','volvo',inplace=True)
auto['Make'].replace('Peugot','peugot',inplace=True)
auto['Make'].replace('vol00112ov','volvo',inplace=True)
auto['No.of Doors'].replace('fourR','four',inplace=True)
auto['Aspiration'].replace('turrrrbo','turbo',inplace=True)
auto['Engine Location'].replace('FRONT','front',inplace=True)
auto['Engine Location'].replace('Front','front',inplace=True)
auto['Engine Location'].replace('Rear','rear',inplace=True)
auto['Engine Location'].replace('REAR','rear',inplace=True)
auto['Fuel Type'].replace('diesel','Diesel',inplace=True)
auto['Fuel Type'].replace('gas','Gas',inplace=True)
auto['Fuel Type'].replace('gas ','Gas',inplace=True)
auto['No. of Cylinders'].replace('Four','four',inplace=True)
auto['No. of Cylinders'].replace('four ','four',inplace=True)
auto['No. of Cylinders'].replace('five','five',inplace=True)
auto['No. of Cylinders'].replace('four','four',inplace=True)
auto['No. of Cylinders'].replace('five ','five',inplace=True)
auto['Body Style'].replace('sedan ','sedan',inplace=True)
auto['No. of Cylinders']=auto['No. of Cylinders'].str.strip()
auto['Make']=auto['Make'].str.strip()
auto['Drive Wheels']=auto['Drive Wheels'].str.strip()
auto['Body Style'].replace('Sedan','sedan',inplace=True)
auto['Body Style'].replace('Wagon','wagon',inplace=True)
print(auto)
```

```python
print(auto.describe())
```

```python
#To check number of missing values in each of the coloumn
auto.isnull().sum()
```

```python
#Missing value treatment for Normalized Losses grouped by the make of the car and if not available then using Body Style
auto['Normalized Losses'] = auto['Normalized Losses'].fillna(auto.groupby('Make')['Normalized Losses'].transform('mean'))
auto['Normalized Losses'] = auto['Normalized Losses'].fillna(auto.groupby('Body Style')['Normalized Losses'].transform('mean'))
auto
```

```python
#Missing value treatment for Price, Bore, Stroke, Horsepower, Peak RPM grouped by the make of the car using mean values.
auto['Price'] = auto['Price'].fillna(auto.groupby('Make')['Price'].transform('mean'))
auto['Bore'] = auto['Bore'].fillna(auto.groupby('Make')['Bore'].transform('mean'))
auto['Stroke'] = auto['Stroke'].fillna(auto.groupby('Make')['Stroke'].transform('mean'))
auto['Horsepower'] = auto['Horsepower'].fillna(auto.groupby('Make')['Horsepower'].transform('mean'))
auto['Peak RPM'] = auto['Peak RPM'].fillna(auto.groupby('Make')['Peak RPM'].transform('mean'))
auto.isnull().sum()
```
